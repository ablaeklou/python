```python
import csv
import matplotlib.pyplot as plt

# Function to read CSV into a list of dictionaries
def read_csv_to_dict_list(file_path):
    data_list = []
    with open(file_path, 'r') as csv_file:
        csv_reader = csv.DictReader(csv_file)
        for row in csv_reader:
            data_list.append(row)
    return data_list

# File path for the Iris dataset
iris_csv_path = "iris.csv"

# Read CSV into a list of dictionaries
iris_data_list = read_csv_to_dict_list(iris_csv_path)

# Display the first few rows of the dataset
print("First Five Rows:")
for row in iris_data_list[:5]:
    print(row)

# Display specific rows
print("Rows 52 to 57:")
for row in iris_data_list[51:57]:
    print(row)

print("Rows 118 to 123:")
for row in iris_data_list[117:123]:
    print(row)


# Display the count of each variety
variety_counts = {}
for row in iris_data_list:
    variety = row["variety"]
    if variety in variety_counts:
        variety_counts[variety] += 1
    else:
        variety_counts[variety] = 1
print("\nVariety Counts:")
print(variety_counts)

# Plot histogram for sepal length
sepal_length_values = [float(row["sepal.length"]) for row in iris_data_list if row["sepal.length"].replace('.', '', 1).isdigit()]
plt.hist(sepal_length_values, bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Frequency')
plt.show()

# Plot boxplot for sepal width
sepal_width_values = [float(row["sepal.width"]) for row in iris_data_list if row["sepal.width"].replace('.', '', 1).isdigit()]
plt.hist(sepal_width_values, bins=20, color='skyblue', edgecolor='black')
plt.title('histogram of Sepal Width')
plt.xlabel('Sepal Width')
plt.ylabel('Value')
plt.show()


# Plot petal length
petal_length_values = [float(row["petal.length"]) for row in iris_data_list if row["petal.length"].replace('.', '', 1).isdigit()]
plt.hist(petal_length_values, bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Petal Length')
plt.xlabel('Petal Length')
plt.ylabel('Frequency')
plt.show()

# Plot petal width
petal_width_values = [float(row["petal.width"]) for row in iris_data_list if row["petal.width"].replace('.', '', 1).isdigit()]
plt.hist(petal_width_values, bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of Petal Width')
plt.xlabel('Petal Width')
plt.ylabel('Frequency')
plt.show()
```
